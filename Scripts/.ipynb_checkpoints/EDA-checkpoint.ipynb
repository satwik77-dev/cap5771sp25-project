# EDA for Dataset 1: Mental Health Counseling Conversations

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import nltk
from nltk.tokenize import word_tokenize
from collections import Counter
```

```python
# Load the cleaned dataset
df1 = pd.read_csv("cleaned_counseling_conversations.csv")

# Display dataset information
df1.info()
# Display first few 15 rows
df1.head(15)
```

```python
df1 = df1[['Cleaned_Context', 'Cleaned_Response']]
```

```python
# Display dataset information
df1.info()
# Display first few 15 rows
df1.head(15)
```

```python
# Check for missing values
print("Missing values before handling:")
print(df1.isnull().sum())
```

```python
#Handling Missing Values
df1 = df1.dropna(subset=['Cleaned_Response'])

#Missing values after Handling
print("Missing values after Handling:")
print(df1.isnull().sum())
```

# Descriptive Statistics

## Analyze Message Length (Word Count)

```python
# Add new columns for word count
df1['Context_Word_Count'] = df1['Cleaned_Context'].apply(lambda x: len(str(x).split()))
df1['Response_Word_Count'] = df1['Cleaned_Response'].apply(lambda x: len(str(x).split()))

# Summary statistics
print("Descriptive Statistics for Word Count:")
print(df1[['Context_Word_Count', 'Response_Word_Count']].describe())
```

```python
# Plot distribution of word counts
plt.figure(figsize=(12, 5))

#Context Word Count
plt.subplot(1, 2, 1)
sns.histplot(df1['Context_Word_Count'], bins=30, kde=True, color='blue')
plt.title("Distribution of Question Length (Word Count)")
plt.xlabel("Number of Words")
plt.ylabel("Frequency")

#Response Word Count
plt.subplot(1, 2, 2)
sns.histplot(df1['Response_Word_Count'], bins=30, kde=True, color='green')
plt.title("Distribution of Response Length (Word Count)")
plt.xlabel("Number of Words")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

Context Word Count
- This histogram is Right-skewed, means that most of the questions contain a small number of words with the peak around 10–20 words suggests that most questions are short.
- Few questions have more than 100 words, with very rare cases extending beyond 200.

Response Word Count
- This histogram is also Right-skewed, but the distribution is wider compared to questions, with the peak around 50 words suggests that most responses are longer than questions.
- Some responses contain over 200 words, meaning that some responses can be quite detailed.

# Most Frequent Words in User Queries & Therapist Responses

```python
#User Queries

# Tokenize words for frequency analysis
nltk.download('punkt')

# Get most common words in user questions
question_words = " ".join(df1['Cleaned_Context']).split()
question_freq = Counter(question_words).most_common(20)

# Convert to DataFrame for visualization
question_df = pd.DataFrame(question_freq, columns=['Word', 'Frequency'])

# Plot top words in user questions
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.barplot(data=question_df, x='Frequency', y='Word', color='blue')
plt.title("Top 20 Most Common Words in User Queries")
plt.xlabel("Frequency")
plt.ylabel("Words")
plt.tight_layout()
plt.show()
```

- The word "im" (I'm) appears most frequently, suggesting that users often describe personal feelings or situations.
- Other high-frequency words include "feel", "dont", "like", "know", "get", and "want", which indicate that users commonly express emotions, uncertainties, and desires.
- Words like "love", "relationship", "friend" and "think" suggest that topics related to relationships and emotions are common.
- "year" and "time" could indicate discussions about past events or time-related concerns.

```python
# Therapist Responses

# Get most common words in therapist responses
response_words = " ".join(df1['Cleaned_Response']).split()
response_freq = Counter(response_words).most_common(20)

# Convert to DataFrame for visualization
response_df = pd.DataFrame(response_freq, columns=['Word', 'Frequency'])

# Plot top words in Therapist Resposes
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 2)
sns.barplot(data=response_df, x='Frequency', y='Word', color='green')
plt.title("Top 20 Most Common Words in Therapist Responses")
plt.xlabel("Frequency")
plt.ylabel("Words")
plt.tight_layout()
plt.show()
```

- "feel" is the most frequently used word, indicating that therapists validate and acknowledge emotions in their responses.
- "may", "help", "would" and "relationship" are also common, suggesting a focus on guidance, possibility, and relationship-related advice.
- The use of modal verbs like "may" and "would" suggests that therapists often offer possibilities, suggestions, or encouragement rather than giving direct commands.

# Word Cloud for User Queries & Therapist Responses

```python
# Generate word cloud for user queries
wordcloud_questions = WordCloud(width=800, height=400, background_color='white').generate(" ".join(df1['Cleaned_Context']))

# Plot word clouds
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.imshow(wordcloud_questions, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud of User Queries")
plt.show()
```

- Words like "don't", "know", "feel", "time", "issue" and "address" appear prominently, suggesting that users often express confusion, emotions, time-related concerns, and specific issues they want to address.
- "relationship", "love" and "family" suggest that many users discuss interpersonal relationships and possibly seek advice on them.

```python
# Generate word cloud for Therapist Responses
wordcloud_responses = WordCloud(width=800, height=400, background_color='white').generate(" ".join(df1['Cleaned_Response']))

# Plot word clouds
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 2)
plt.imshow(wordcloud_responses, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud of Therapist Responses")
plt.show()
```

- "relationship" is the most prominent word, indicating that many therapist responses involve discussions about relationships (romantic, family, or social).
- "feel", "feeling" and "need" suggest that therapists focus on emotional validation and addressing user needs.
- The use of modal verbs like "may" and "want" suggests that therapists often provide possibilities, encourage reflection, and avoid imposing solutions.

# Identify Therapy Techniques in Therapist Responses

```python
# Define therapy-related keywords
therapy_keywords = ["cbt", "mindfulness", "meditation", "deep breathing", "coping", "therapy", "counseling"]

# Count occurrences in responses
df1['Contains_Therapy_Method'] = df1['Cleaned_Response'].apply(lambda x: any(word in x for word in therapy_keywords))

# Plot therapy technique presence
plt.figure(figsize=(6, 4))
sns.countplot(x=df1['Contains_Therapy_Method'], hue=df1['Contains_Therapy_Method'], palette=["red", "green"], legend=False)
plt.xticks([0, 1], ["No Therapy Method", "Includes Therapy Method"])
plt.title("Presence of Therapy Methods in Responses")
plt.show()
```

- Most Responses Do Not Explicitly Mention Therapy Techniques. Instead, therapists may be providing general guidance, emotional support, or alternative advice.
- A Smaller but Noticeable Portion of Responses Include Therapy Methods. This suggests that while therapy techniques are used, they are not always explicitly stated in every response.

# EDA for Dataset 2: Mental Health FAQ for Chatbot

```python
# Load the cleaned dataset
df2 = pd.read_csv("cleaned_mental_health_faq.csv")

# Display dataset information
df2.info()

# Display first few rows
df2.head(10)
```

```python
df2 = df2[['Cleaned_Questions', 'Cleaned_Answers']]
# Display dataset information
df2.info()
# Display first few 15 rows
df2.head(15)
```

```python
# Check for missing values
print("Missing values before handling:")
print(df2.isnull().sum())
```

Therefore, There are no Missing values in this Cleaned Dataset.

## Descriptive Statistics

Analyze Question & Answer Length (Word Count)

```python
# Add new columns for word count
df2.loc[:, 'Question_Word_Count'] = df2['Cleaned_Questions'].apply(lambda x: len(str(x).split()))
df2.loc[:, 'Answer_Word_Count'] = df2['Cleaned_Answers'].apply(lambda x: len(str(x).split()))

# Summary statistics
print("Descriptive Statistics for Word Count:")
print(df2[['Question_Word_Count', 'Answer_Word_Count']].describe())
```

```python
# Plot distribution of FAQ Question Length
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df2['Question_Word_Count'], bins=20, kde=True, color='blue')
plt.title("Distribution of FAQ Question Length (Word Count)")
plt.xlabel("Number of Words")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

- This Histogram plot is Right-skewed, meaning most FAQ questions are short, but a few are longer. Where peak occurs between 3-6 words, indicating that most questions are concise and very few questions exceed 10 words, and only a rare case extends beyond 20 words.

```python
# Plot distribution of FAQ Answer Length
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 2)
sns.histplot(df2['Answer_Word_Count'], bins=20, kde=True, color='green')
plt.title("Distribution of FAQ Answer Length (Word Count)")
plt.xlabel("Number of Words")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

- This Histogram distribution is Right-skewed, meaning most answers are short to medium-length, but a few are very long. With a significant number of responses have less than 100 words, with a peak around 50 words.Some answers also extend well beyond 300–700 words, indicating that some responses are highly detailed.
- While deploying chatbot, responses should be optimized based on query type:
Simple yes/no or definition-based FAQs should retrieve concise answers.
How-to or guidance-based FAQs should fetch longer, detailed responses.

## Most Frequent Words in FAQ Questions & Answers

```python
# Get most common words in FAQ questions
question_words = " ".join(df2['Cleaned_Questions']).split()
question_freq = Counter(question_words).most_common(20)
# Convert to DataFrame for visualization
question_df = pd.DataFrame(question_freq, columns=['Word', 'Frequency'])

# Plot top words in FAQ questions
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.barplot(data=question_df, x='Frequency', y='Word', color='blue')
plt.title("Top 20 Most Common Words in FAQ Questions")
plt.xlabel("Frequency")
plt.ylabel("Words")
plt.tight_layout()
plt.show()
```

- Words like "mental", "health", "disorder", "illness", "medication" and "care" indicate that most FAQ questions revolve around mental health topics.
- Common words like "find", "help", "whats", "difference", "know" and "use" suggest that many questions are information-seeking.

```python
# Get most common words in FAQ answers
answer_words = " ".join(df2['Cleaned_Answers']).split()
answer_freq = Counter(answer_words).most_common(20)

# Convert to DataFrame for visualization
answer_df = pd.DataFrame(answer_freq, columns=['Word', 'Frequency'])

# Plot top words in FAQ answers
plt.subplot(1, 2, 2)
sns.barplot(data=answer_df, x='Frequency', y='Word', color='green')
plt.title("Top 20 Most Common Words in FAQ Answers")
plt.xlabel("Frequency")
plt.ylabel("Words")
plt.tight_layout()
plt.show()
```

- The most frequent words, "health" and "mental" confirm that the responses primarily focus on Mental health topics.
- The words "people" "experience" "family" and "feel" suggest that responses address human experiences, emotional support, and family involvement.
- Words like "help" "support" "treatment" and "doctor" suggest that many answers offer Guidance.

## Word Cloud for FAQ Questions & Answers

```python
# Generate word cloud for FAQ questions
wordcloud_questions = WordCloud(width=800, height=400, background_color='white').generate(" ".join(df2['Cleaned_Questions']))

# Plotting Word Cloud for FAQ Questions
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.imshow(wordcloud_questions, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud of FAQ Questions")
plt.show()
```

```python
# Generate word cloud for FAQ Answers
wordcloud_answers = WordCloud(width=800, height=400, background_color='white').generate(" ".join(df2['Cleaned_Answers']))

# Plot Word cloud for FAQ Answers
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 2)
plt.imshow(wordcloud_answers, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud of FAQ Answers")
plt.show()
```

- Most frequent word is 'may' which shows that FAQ Answers mostly include different mitigation scenarios related to the FAQ Answers.
- The most prominent words include "mental", "health", "treatment", "symptom", "problem", "help" and "medication". This suggests that FAQ answers heavily focus on mental health conditions, treatment options, and medical guidance.
- Words like "feel", "experience", "need", "people", "important" and "support" indicate that responses also address emotional well-being, social support, and personal experiences. 

## Identifying Top Mental Health Topics

```python
# Define mental health keywords
mental_health_keywords = ["anxiety", "depression", "stress", "therapy", "cbt", "medication", "counseling", "support"]

# Count occurrences in FAQ questions
df2['Contains_Mental_Health_Topic'] = df2['Cleaned_Questions'].apply(lambda x: any(word in x for word in mental_health_keywords))
# Plot topic presence
plt.figure(figsize=(6, 4))
sns.countplot(x=df2['Contains_Mental_Health_Topic'], hue=df2['Contains_Mental_Health_Topic'], palette=["red", "green"], legend=False)
plt.xticks([0, 1], ["No Mental Health Topic", "Includes Mental Health Topic"])
plt.title("Presence of Mental Health Topics in FAQ Questions")
plt.show()
```

- Majority of Questions Do Not Explicitly Mention Mental Health Terms. This may imply that most Users may be asking general well-being or lifestyle-related questions that indirectly relate to mental health.
- A Smaller Portion of Questions Explicitly Discuss Mental Health Topics.

# EDA for Dataset 3: Sentiment Analysis for Mental Health

```python
# Load the cleaned sentiment analysis dataset
df3 = pd.read_csv("cleaned_sentiment_analysis.csv")

# Display dataset information
df3.info()

# Display first few 15 rows
df3.head(15)
```

```python
#Removing 'statement' column in Original Cleaned dataset
df3 = df3[['Cleaned_Statement', 'status', 'Encoded_Status']]
```

```python
#Handling Missing Values
df3 = df3.dropna()
```

```python
#Verify whether Changes are made or not
df3.info()
df3.head(15)
```

# Descriptive Statistics

## Analyze Statement Length (Word Count)

```python
# Add new column for word count
df3['Statement_Word_Count'] = df3['Cleaned_Statement'].apply(lambda x: len(str(x).split()))

# Summary statistics
print("Descriptive Statistics for Word Count:")
print(df3[['Statement_Word_Count']].describe())
```

```python
# Plot distribution of statement word counts
plt.figure(figsize=(8, 5))
sns.histplot(df3['Statement_Word_Count'], bins=30, kde=True, color='purple')
plt.title("Distribution of Statement Length (Word Count)")
plt.xlabel("Number of Words")
plt.ylabel("Frequency")
plt.show()
```

- This Histogram is a Highly Right-Skewed Distribution with majority of statements are very short, clustering towards the left (near 0-50 words) with a very few outliers have extremely long word counts (some exceeding 2000+ words)
- The peak at low word counts suggests that most statements are direct or single-sentence responses.

# Distribution of Mental Health Categories

```python
# Count of each mental health status
plt.figure(figsize=(8, 5))
sns.countplot(y=df3['status'], order=df3['status'].value_counts().index, hue=df3['status'], legend=False)
plt.title("Distribution of Mental Health Categories")
plt.xlabel("Count")
plt.ylabel("Mental Health Condition")
plt.show()
```

- The Largest category is "Normal", meaning most statements do not indicate significant mental health distress. "Depression" is the second most common category, indicating a significant number of statements discuss depressive thoughts and emotions.
- The "Suicidal" category also has a notable presence, which is critical from a mental health intervention perspective.
- "Anxiety" and "Stress" appear much less frequently compared to Depression and Suicidal thoughts. This could mean that people tend to describe depressive symptoms more explicitly than general anxiety or stress.

# Word Cloud for Different Mental Health Conditions

```python
# Generate word clouds for different mental health conditions
plt.figure(figsize=(14, 8))
for i, category in enumerate(df3['status'].unique(), 1):
    text = " ".join(df3[df3['status'] == category]['Cleaned_Statement'])
    wordcloud = WordCloud(width=500, height=300, background_color='white').generate(text)
    plt.subplot(3, 3, i)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title(f"Word Cloud - {category}")
plt.tight_layout()
plt.show()
```

- Depression & Suicidal categories contain words such as "feel", "life", "think", "thing", "time", "even" and "know" reflecting deep thought, emotions, and existential concerns.
- Stress & Anxiety discussions are more situational and physical, often tied to work, symptoms, and time constraints with words such as "feel", "anxiety", "stress", "time", "symptom", "work" and "life."
- Normal statements focus on neutral, day-to-day topics with a mix of casual expressions and desires.

# Relationship Between Statement Length & Mental Health Category

```python
# BoxPlot for Relationship Between Statement Length & Mental Health Category
plt.figure(figsize=(10, 6))
sns.boxplot(x=df3['Encoded_Status'], y=df3['Statement_Word_Count'], hue=df3['Encoded_Status'], palette='coolwarm', legend=False)
plt.title("Statement Length Across Mental Health Conditions")
plt.xlabel("Encoded Mental Health Status")
plt.ylabel("Number of Words in Statement")
plt.show()
```

- The boxes (IQR) are compressed near the bottom, indicating that most statements are Short (low word count) across all mental health categories and also Medians (horizontal lines inside the boxes) for all categories are close to zero supporting this.
- Each category has outliers extending upwards from 1000 to 3500+ words suggesting that some individuals express their thoughts in much longer, detailed statements.

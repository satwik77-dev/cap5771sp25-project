### On Dataset 3 : Sentiment Analysis for Mental Health 

Feature: Has_Severe_Keyword

- Flag if the statement contains high-risk keywords (e.g., die, suicide, worthless, kill, hopeless) which is useful for alerting support systems.

```python
#Importing Library
import pandas as pd

# Load Dataset 3 : Sentiment Analysis for Mental Health
df_sentiment = pd.read_csv("/Users/satwik/Documents/GitHub/cap5771sp25-project/Data/Cleaned Data/Cleaned_Sentiment_Analysis.csv")
```

```python
# Define severe keywords set
severe_keywords = {'die', 'suicide', 'worthless', 'kill', 'hopeless', 'pain', 'end', 'cut', 'hurt', 'alone'}

# Function to detect presence of any severe keyword
def has_severe_keyword(text):
    if isinstance(text, str):
        tokens = text.lower().split()
        return int(any(word in severe_keywords for word in tokens))
    else:
        return 0
# Apply function to Cleaned_Statement column
df_sentiment['Has_Severe_Keyword'] = df_sentiment['Cleaned_Statement'].apply(has_severe_keyword)
```
